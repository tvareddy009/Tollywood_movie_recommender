```python
movies=movies.drop(columns=['Runtime','No.of.Ratings','Unnamed: 0'])
```

```python
movies['Overview']=movies['Overview']+movies['Genre']+movies['Certificate']+movies['Rating'];
```

```python
movies.iloc[0]['Overview']
```

```python
movies.drop(columns=['Genre','Certificate','Rating'])
movies.reset_index(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,stop_words='english')
```

```python
v1=cv.fit_transform(movies['Overview']).toarray()
```

```python
v1.shape
```

```python
movies.iloc[0]['Overview']
```

```python
v1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(v1)
```

```python
sim
```

```python
movies[movies['Movie']=='1 - Nenokkadine'].index[0]
```

```python
def recommend(mov):
   index=movies[movies['Movie']==mov].index[0]
   distances=sorted(list(enumerate(sim[index])),reverse=True,key = lambda x: x[1:6])
   for i in distances[1:6]:
        print(movies.iloc[i[0]].Movie)
```

```python
recommend('Baahubali 2: The Conclusion')
```
